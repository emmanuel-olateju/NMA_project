### activity and CCA results

#### function

```python
def bg_activity_CCA_result_plot(result,i):
    df=pd.DataFrame(columns=list(result.keys()))
    for c in df.columns:
        df[c]=result[c].param['window_shift_CCA'][i,i,:]
    plt.figure(figsize=(24,10))
    sns.heatmap(df.T,annot=True)
```
